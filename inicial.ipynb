### Case - Cancelamento de Clientes

Você foi contratado por uma empresa com mais de 300 mil clientes para um projeto de Dados. Recentemente a empresa percebeu que da sua base total de clientes, a maioria são clientes inativos, ou seja, que já cancelaram o serviço.

Precisando melhorar seus resultados ela quer conseguir entender os principais motivos desses cancelamentos e quais as ações mais eficientes para reduzir esse número.




(Utilize a base de dados "cancelamentos_amostra.csv")

```python
import pandas 

# comando para ler um arquivo:  pandas.read_[tipo do arquivo]("[nome do arquivo]")
# neste caso o comando é: pandas.read_csv("cancelamentos.csv")

tabela = pandas.read_csv("cancelamentos_amostra.csv")

# comando para mostrar tabelas de maneira "bonitinha": display([base de dados, em questão, que foi lida com pandas.read])
# neste caso o comando é: display(tabela)

display(tabela)

#site para tu pegar bases de dados e treinar situações como esta: https://www.kaggle.com/


```

```python
#obs_dica_técnica: Informações que não te ajudam, informações vazias e informações em formatos errados te atrapalham.
#Neste caso, a coluna "CostumerID" está te atrapalhando.


#Comando "[base de dados lida anteriormente].drop([o que voce quer jogar fora])" : joga fora alguma coisa da tabela
#Neste caso: tabela.drop(columns="CustomerID")


tabela = tabela.drop(columns="CustomerID")


display(tabela)


#comando que exibe informações da base de dados: [base de dados lida anteriormente].info()
#Neste caso tabela.info()

display(tabela.info())






```

```python
#comando "[Base de dados lida anteriormente].dropna()" : elimina linhas vazias da sua base de dados lida (da sua tabela)
#neste caso: tabela.dropna()

#comando "[Base de dados lida anteriormente].drop_duplicates()"
#neste caso: tabela.drop_duplicates()

tabela = tabela.dropna()

display(tabela.info())



```

```python
###############     Entendendo quantos clientes cancelaram:       #####################

#Ao analisar a tabela exibida com display(tabela); 
#considere, na coluna "canselou" -ou a última coluna-, o numero 0.0 como uma pessoa que não cancelou e 1.0 como alguém que cancelou


#comando para selecionar uma coluna [Base de dados lida anteriormente]["Nome da coluna literalmente entre aspas seguidas de colchetes"]
#Neste caso: tabela["cancelou"]

#comando para contar valores de uma coluna: [coluna selecionada].value_counts()
#neste caso: tabela["cancelou"].value_counts()

display(tabela["cancelou"].value_counts())

#calculando em percentual
#so pegar a estrutura anterior e colocar normalize=True dentro do parenteses de value_counts()


#codigo a ser executado: display(tabela["cancelou"].value_counts(normalize=True))


display(tabela["cancelou"].value_counts(normalize=True).map("{:.1%}".format))

#obs: o .map("{:.1%}".format) é só para formatar os valores que estão em decimais para porcentagem, apenas para ficar estéticamente mais agradável

 


```

```python
##################      Esboçando em gráficos      #####################

#Obs: Para isso será utilizada a biblioteca plotly

import plotly.express as px  # <-- Sempre importe a biblioteca plotly assim

#comando para criar gráficos: 
# px.[tipo de gráfico que você quer criar]([Base de dados lida anteriormente], x="coluna que voce quer", color = "coluna/atributo que voce quer distinguir os dados por cores")

grafico = px.histogram(tabela, x="duracao_contrato", color="cancelou", text_auto="True")

#comando que exibe o gráfico: [Gráfico feito anteriormente].show()

grafico.show()




```

```python
#Analisando os cancelamentos de cada coluna:


for coluna in tabela.columns:
    grafico = px.histogram(tabela, x=coluna, color="cancelou", text_auto="True")

    grafico.show()
```

```python
### Analisando os gráficos ###

#Todo mundo do contrato mensal canselou o serviço
    #Vamos dar desconto ao contrato Anual e Trimestral

#Ligações_callcenter acima de 4 resulta em cancelamento
    #Se o cliente ligar 3 vezes, alerta vermelho!

#Atraso no pagamento acima de 20 dias resulta em cancelamento
    #Se o cliente atrasar 15 dias, alerta vermelho!






### Estimando como os numeros vão mudar caso cada um dos três ultimos problemas sejam resolvidos ###


print("Essa é a tabela, com as taxas de cancelamento, antes de resolvermos os problemas x, y e z: \n 1.0 -> representa os clientes que cancelaram \n 0.0 -> representa os clientes que não cancelaram \n")

display(tabela["cancelou"].value_counts(normalize="True"))


#Para isso, faremos um filtro que desconsiderará os dados ruins

#duracao_contrato -> diferente de Monthly (Pois queremos tirar as pessoas da assinatura mensal)
condicao = tabela["duracao_contrato"] != "Monthly"
tabela = tabela[condicao]


#ligacoes_callcenter -> menores ou iguais a 4 (Pois não queremos que os clientes liguem para o callcenter mais de 4 vezes)
condicao = tabela["ligacoes_callcenter"] <= 4
tabela = tabela[condicao]


#dias_atraso -> menores ou iguais a 20 (Pois não queremos que os clientes atrasem mais que 20 dias)
condicao = tabela["dias_atraso"] <= 20
tabela = tabela[condicao]

print(" \n \n \n \n")

print("Essa é a tabela, com as taxas de cancelamento, depois de resolvermos os problemas x, y e z: \n 1.0 -> representa os clientes que cancelaram \n 0.0 -> representa os clientes que não cancelaram \n")
display(tabela["cancelou"].value_counts(normalize="True"))

#porcentagem

```
